```python
data.payer_inn = pd.to_numeric(data.payer_inn, errors='coerse')
```

```python
data.summa = pd.to_numeric(data.summa, errors='coerse')
```

```python
data.shape
```

```python
data = data[~data.summa.isnull()]
```

```python
data = data[data.payer_inn > 1e6]
```

```python
data.payee_inn = pd.to_numeric(data.payee_inn, errors='coerse')
```

```python
data = data[data.payee_inn > 1e6]
```

```python
data_filter = data.groupby('payer_inn').filter(lambda x :  (len(x) > 100) and (x.summa.sum()>1e8))
```

#### Транзакционные данне с января 2015 по январь 2019. 
#### 12789 уникальных ИНН клиентов с количеством транзакций более 100 за указанный период и суммой транзакций более 1 млн. рублей.

```python
data_filter = data_filter.groupby(['payer_inn','payee_inn']).summa.sum()
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
counts = data_filter.groupby('payer_inn').count()
counts = counts.clip(0, np.percentile(counts, 99))
counts.hist(figsize=(20, 10),xlabelsize=15,ylabelsize=15,bins=50)
plt.title('Активность клиентов',fontsize=20)
plt.ylabel('Количество клиентов', fontsize=20)
plt.xlabel('Количество уникальных физических лиц с которыми клиент взаимодействует', fontsize=20)
```

```python
percentage_crd = data_filter.groupby('payer_inn').apply(lambda x: x/x.sum())
```

```python
percentage_crd = percentage_crd[percentage_crd > 0.05]
```

```python
a, bins = pd.qcut(percentage_crd, q=10, labels=range(10),retbins=True)
```

```python
bins
```

```python
import networkx as nx
```

```python
a = a.reset_index()
```

```python
list_of_relations = [string.values  for i, string in a.iterrows()]
```

```python
g = nx.Graph()
```

```python
g.add_weighted_edges_from(list_of_relations[:400])
```

```python
plt.figure(figsize=(30,30))
pos=nx.spring_layout(g, k=2) # positions for all nodes

nx.draw_networkx_edge_labels(g, pos, font_size=10)
# nodes
nx.draw_networkx_nodes(g, pos, node_size=100,)

# edges
nx.draw_networkx_edges(g,pos,width=0.1)

## Двойной щелчек мыши по картинке для увеличения.
```
